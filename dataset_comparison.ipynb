```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)
```

```python
webscrape_df = pd.read_csv('Dataset2/webscrape.csv')
webscrape_df.head()
```

```python
players_df = pd.read_csv('Data/players.csv')
players_df.head()
```

```python
bowlers = set(webscrape_df['Bowler'].unique())
batsmen = set(webscrape_df['Batsman'].unique())
players = set(players_df['Name'].unique())

bowlers_not_in_players = bowlers - players
batsmen_not_in_players = batsmen - players

players_not_in_webscrape = players - (bowlers | batsmen)

print(f"Total Bowlers: {len(bowlers)}, Not in Players: {len(bowlers_not_in_players)}")
print(f"Total Batsmen: {len(batsmen)}, Not in Players: {len(batsmen_not_in_players)}")
print(f"Total Players: {len(players)}, Not in Bowler or Batsman: {len(players_not_in_webscrape)}")
```
