```python
df['isWicket'] = df['isWicket'].astype(int)
```

```python
plt.figure(figsize=(20, 5))
sns.boxplot(x=df['shotType'], y=df['isWicket'])
plt.xlabel('Shot Type')
plt.ylabel('Is Wicket')
plt.title('Shot Type vs Is Wicket')
plt.xticks(rotation=90)
plt.show()
```

```python
heatmap_data = df.pivot_table(index='pitchLine', columns='shotType', values='isWicket')

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', fmt='g')
plt.xlabel("Shot Type")
plt.ylabel("Pitch Line")
plt.title("Heatmap of Shot Type vs Pitch Line (Hue: Wickets)")
plt.show()
```

```python
heatmap_data = df.pivot_table(index='pitchLength', columns='shotType', values='isWicket')

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', fmt='g')
plt.xlabel("Shot Type")
plt.ylabel("Pitch Length")
plt.title("Heatmap of Shot Type vs Pitch Length (Hue: Wickets)")
plt.show()
```
